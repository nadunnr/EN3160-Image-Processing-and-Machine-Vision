```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im_path = r"D:\Academics\EN3160 - Image Processing and Machine Vision\images\emma.jpg"

im = cv.imread(r"D:\Academics\EN3160 - Image Processing and Machine Vision\images\emma.jpg")

fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
ax.xaxis.set_ticks_position('top')
plt.show()
```

```python
print("image shape: ", im.shape)
print("image type: ", im.dtype)
print("image size: ", im.size)
```

```python
#incread=sing the brightness

im1 = cv.imread(im_path, cv.IMREAD_GRAYSCALE)
im2 = cv.add(im1, 100)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(im1, cmap='gray', vmin=0, vmax=255)
ax[1].imshow(im2, cmap='gray', vmin=0, vmax=255)
plt.show()
```
